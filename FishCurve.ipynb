```python
#Calculation
from google.colab import drive
drive.mount('/content/drive')
import cv2 as cv
import numpy as np
from numpy import sin
from matplotlib import pyplot
import matplotlib.pyplot as plt
from sklearn import preprocessing
from scipy.optimize import curve_fit
from google.colab.patches import cv2_imshow

img_1 = cv.imread('/content/drive/MyDrive/Java/Python/Project3/Images/Firstfishimage.png')
gray = cv.cvtColor(img_1,cv.COLOR_BGR2GRAY)

corners = cv.goodFeaturesToTrack(gray, 600, 0.01, 10)                          
corners = np.int0(corners)

def poly_model6(x, a, b, c, d):
	return a * sin(b - x) + c * x**2 + d

 
def poly_model5(x,a,b, c, d, e, f, g):
  y = (a * x) + (b * x**2) + (c * x**3) + (d * x**4) + (e * x**5) + (f * x **6) + g
  return y

def poly_model3(x,a,b, c, d, e, f):
  y = (a * x) + (b * x**2) + (c * x**3) + (d * x**4) + (e * x**5) + f
  return y

def poly_model2(x,a,b, c, d):
  y = (a * x) + (b * x**2) + (c * x**3) + d
  return y

def poly_model1(x,a,b, c):
  y = (a * x) + (b * x**2) + c 
  return y


  
def poly_model4(x,a,b, c, d, e):
  y = (a * x) + (b * x**2) + (c * x**3) + (d * x**4) + e
  return y



x_axis = []
y_axis = []
for i in corners:
    x, y = i.ravel()
    x_axis.append(x)
    y_axis.append(y)
    cv.circle(img_1, (x,y), 3, (255, 0, 255), 1)
#Normalize

#function call
popt, pcov = curve_fit(poly_model2, x_axis, y_axis)

print(popt)
plt.scatter(x_axis, y_axis, s=1)
xstart = min(x_axis)
xstop = max(x_axis)
increment = 1
print(xstart)
print(xstop)
xmodel = np.arange(xstart, xstop, increment)
ymodel = poly_model2(xmodel, *popt)
plt.plot(xmodel,ymodel, '--', color='blue')
pyplot.show()
cv2_imshow(img_1)
print("x is :",x_axis)
print("y is :",y_axis)
```
